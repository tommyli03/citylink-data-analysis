```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

df = pd.read_csv("../data/metr-la-zero-removed.csv", index_col=0)
df.index = pd.to_datetime(df.index)
```

```python
def is_bimodal(data, alpha = 0.05):
    kde_values = sns.kdeplot(data, bw_adjust = 2).get_lines()[0].get_data()
    local_maxima_indices = argrelextrema(kde_values[1], np.less)
    valley = kde_values[0][local_maxima_indices]
    threshold = 0
    proportion = 0
    for v in valley:
        thresh = v
        prop = sum(data < thresh)/len(data)
        if abs(prop - 0.5) < abs(proportion - 0.5):
            proportion = prop
            threshold = thresh
    if min(proportion, 1 - proportion) > alpha:
        print(f"threshold: {threshold}, prop: {proportion}")
        plt.hist(data, bins = 100)
        plt.axvline(x=threshold, color='red', linestyle='--')
        plt.show()
    plt.clf()
    return min(proportion, 1 - proportion)
```

```python
result = df.apply(is_bimodal)
```

```python
sum(result > 0.05)/len(result)
```

```python
locations = pd.read_csv('../data/sensor_graph/graph_sensor_locations.csv', index_col=1)
locations.index = locations.index.astype(str)
```

```python
result = pd.DataFrame(result, columns=['Proportion'])
```

```python
df = pd.concat([locations, result], axis=1)
```
